```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('marks.csv')

students = data['Name']
subjects = data.columns[1:]

df = data.set_index('Name')

average_marks = df.mean(axis=1)

print("\nIndividual Performance:(Average Marks)")
print(average_marks)

subject_difficulty = df.mean()


rankings = average_marks.sort_values(ascending=False)
first, second, third = rankings.index[:3]
print(f"\n1st Position: {first}")
print(f"2nd Position: {second}")
print(f"3rd Position: {third}")

```

```python

plt.figure(figsize=(12, 6))
plt.bar(rankings.index, rankings.values, color='blue')
plt.xlabel('Students')
plt.ylabel('Average Marks (out of 30)')
plt.title('Overall Performance of Students')
plt.xticks(rotation=45)
plt.show()

```

```python

plt.figure(figsize=(8, 8))
plt.pie(subject_difficulty, labels=subject_difficulty.index, autopct='%1.1f%%', startangle=140)
plt.title('Difficulty Level of Subjects (Lower Mean = Harder)')
plt.show()
```

```python

plt.figure(figsize=(10, 5))
plt.hist(df.values.flatten(), bins=10, color='green', alpha=0.7, edgecolor='black', linewidth=1.2)
plt.xlabel('Marks (out of 30)')
plt.ylabel('Frequency')
plt.title('Distribution of Marks')
plt.show()

```

```python

# Individual student performance (bar chart)
for student in df.index:
    plt.figure(figsize=(8, 5))
    plt.bar(df.columns, df.loc[student], color='purple')
    plt.xlabel('Subjects')
    plt.ylabel('Marks (out of 30)')
    plt.title(f'Performance of {student}')
    plt.ylim(0, 30)
    plt.xticks(rotation=45)
    plt.show()
```
